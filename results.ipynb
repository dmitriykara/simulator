```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['figure.figsize'] = (20,10)
```

```python
params = pd.read_csv("results/params.csv", sep=',')
params
```

```python
requests = pd.read_csv("results/requests.csv", sep=',')
requests
```

```python
print(f"Mean processing time: {requests['Processed time'].mean()}")
print(f"Mean waiting time: {(requests['Processing started'] - requests['ArrivedAt']).mean()}")
```

```python
states = pd.read_csv("results/states.csv", sep=',')
states
```

```python
print(f"Mean number of packages processing: {states['Processing'].mean()}")
```

```python
plt.plot(states['Processing'])
plt.title('Number of requests processing at the moment')
plt.xlabel('Time')
plt.ylabel('Processing')
plt.show();
```

```python
plt.plot(states['Waiting'])
plt.title('Number of requests waiting at the moment')
plt.xlabel('Time')
plt.ylabel('Waiting')
plt.show();
```

```python
plt.plot(states['Average response time'])
plt.title('Average response time')
plt.xlabel('Time')
plt.ylabel('Average response time')
plt.show();
```
